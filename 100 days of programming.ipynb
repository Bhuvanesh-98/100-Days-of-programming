# DAY 1 CHALLENGE (PATTERN PROGRAMS)

```python
for i in range(0,5):
    for j in range(0,6):
        print("]",end=" ")
    print("\r")

print()
for i in range(0,6):
    for j in range(i+1,0,-1):
        print("* ",end="")
    print("\r")
    
print()
for i in range(6,0,-1):
    for j in range(0,i):
        print("V ",end="")
    print("")
    
print()
num=int(input())
for i in range(num,0,-1):
    for j in range(0,num-i):
        print(end=" ")
    for j in range(0,i):
        print("* ",end="")
    print("\r")
    
print()
num = int(input())
for i in range(0,num):
    for j in range(0,num-i-1):
        print(" ",end="")
    for j in range(0,i+1):
        print("*",end=" ")
    print()

print()
num=int(input())
b=2*num-2
for i in range(0,num):
    for j in range(0,b):
        print(" ",end="")
    b-=2
    for j in range(0,i+1):
        print("% ",end="")
    print()
    
print()
num=int(input())
for i in range(num,0,-1):
    for j in range(i-1,0,-1):
        print("@ ",end="")
    print()
```
